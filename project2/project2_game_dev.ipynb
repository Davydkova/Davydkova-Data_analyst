##  Предобработка данных и замена типа данных

```python
# процентное соотношение пропущенных данных
df.isnull().mean()   
```

```python
# количество уникальных значений в столбце с названием игр
len(df['name'].unique())  

# удаление пропущенных значений
 df=df.dropna(subset=['name'])    
```

Всего 2 пропущенных значения в столбце с играми, скорее всего просто сыграла роль человеческого фактора и их забыли заполнить. 

Уникальное количество названия игр слишком велико чтобы мы смогли корректно заполнить пропущенные значения, но и их всего нечего.

Поэтому безболезненно удаляем пропуски в столбце name

```python
#загасить предупреждение о выводе копии DataFrame
pd.set_option('mode.chained_assignment', None) 

#замена типа в столбце year_of_release на целочисленный
df['year_of_release']=df['year_of_release'].astype('Int64') 
  
#замена  пропущенных данных в столбце year_of_release  на 0
df['year_of_release'].fillna(value=0, inplace=True)
```

Данные исследуются с 1980 года по 2016 включительно.

Обработаем пропущенные значения в годах, заменим их нулевыми значениями и далее перезапишем наш датасет без этих значений

```python
# проверка распределения на начилие нулевых значений в годах
df['year_of_release'].hist()
```

### Замена типа данных 

Проверяем тип данных object в столбцах, он строчный , замена не требуется

Год выпуска игр должен быть целочисленным числом, заменили ранее, в процессе обработки. 

В оценке пользователей тип object, логичнее сделать вещественный, но в  столбце содержится аббревиатура 'tbd'-to be done, она требует замены, как отдельная группа (например "-1")

В данном столбце это скорее всего значит то, что пользователь не оставил свою оценку на данный момент, но воспользовался продуктом, и может оставить свою оценку позже.

В количественном отношении этих данных более 14 процентов, для удаления без существенных потерь это много.

```python
df['user_score']=df['user_score'].replace("tbd", -1) #замена группы "tbd" на "-1"
```

```python
df['user_score']=df['user_score'].astype('Float64')  #преобразование типа в столбце user_score
```

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

    Нулевая гипотеза H₀: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
        
    Альтернативная гипотеза H₁: Средние пользовательские рейтинги платформ Xbox One и PC различаются


```python
xone_hyp = df_active[df_active['platform']=='XOne']['user_score']
pc_hyp = df_active[df_active['platform']=='PC']['user_score']
xone_hyp.mean()                 #средний рейтинг пользователя для XOne платформ
```

```python
pc_hyp.mean() #средний рейтинг пользователя для PC платформ
```

```python
# Произведем проверку дисперсии выборок
np.var(xone_hyp)
```

```python
np.var(pc_hyp)
```

```python
#Выполним проверку гипотезы. Будем использовать метод ttest_ind

alpha = 0.05

results = st.ttest_ind(xone_hyp.dropna(), pc_hyp.dropna(), equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Значение p-value равно  8%.

Таким образом, не получилось опровергнуть Нулевую гипотезу. 

То есть, нам нужно раздвинуть доверительный интервал более чем на 8%, чтобы можно было получить различность рейтингов двух платформ.

### Средние пользовательские рейтинги жанров Action и Sports разные

Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

 Пороговое значение alpha=0.05



    Нулевая гипотеза H₀: Средние пользовательские рейтинги жанров Action и Sports одинаковые.
    Альтернативная гипотеза H₁: Средние пользовательские рейтинги жанров Action и Sports различаются.


```python
genre_action_hyp = df_active[df_active['genre']=='Action'] ['user_score']
genre_sports_hyp = df_active[df_active['genre']=='Sports'] ['user_score']
```

```python
genre_action_hyp.mean() #средняя оценка по жанру Action
```

```python
genre_sports_hyp.mean() #средняя оценка по жанру Sports
```

```python
alpha = 0.05

results = st.ttest_ind(genre_action_hyp.dropna(), genre_sports_hyp.dropna(), equal_var=False)

print('p-значение:', results.pvalue)


if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```



 Получив p-value, мы отвергли Нулевую гипотезу. 
    
 У нас практически нет вероятности получить одинакове средние рейнтинги по жанру Action и Sports.
