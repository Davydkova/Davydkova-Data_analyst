 - Проверка на дубликаты автоматическим методом говорит нам о наличии дубликатов в датафрейме.  
 - При ручной проверке на дубликаты мы видим разное время выгрузки сообщений из каналов и описание вакансий, эти данные не могут быть дублями, их не удаляем. Возможно метод определил дубли по данным json выгруженным из колонкам chat, entities.


```python
# cброс полного отображение данных 
pd.reset_option('^display.', silent=True)
#pd.reset_option('all')

# функция для первичной очистки текста от случайных символов, валюту необходимо оставить
def clear_text(text):
    clear_text = re.sub(r'[^а-яА-ЯёЁa-zA-Z$€*.,\d@]', ' ', text)
    clear_text = " ".join(clear_text.split())
    return clear_text

df['text_'] = df['text'].apply(clear_text)
```

#### Вакансии/специализация

```python
df.text_[5].lower()
```

```python
ds = r'#ds|scien|дс|саен|tensoflow|pytorch|machine|learning|vision|nlp'
da = r'analyst|analysis|аналитик/w|business intelligence|биай|tableau|powerbi|#da|A/B'
de = r'engineer|инженер|spark|airflow|hadooop|scala|#de|etl|linux|архитект[оy]р[аы]|хранилище|разработчик|desktop'
```

```python
# пометим True или False там где у нас совпадения

df['ds'] = df['text_'].apply(lambda x:True if re.search(ds,x.lower()) else False) 
df['da'] = df['text_'].apply(lambda x:True if re.search(da,x.lower()) else False) 
df['de'] = df['text_'].apply(lambda x:True if re.search(de,x.lower()) else False)
```

 - Некоторые вакансии по специалистам могут пересекать в себе релевантный стек инструментов. Например, DA также используют в своей работе etl, airlow. То есть данная вакансия может релевантна к рассмотрению сразу нескольким дата специалистам. В данном анализе она попадет в фильтр DA и DE , далее возможна корректировка и выбор вакансии.

```python
grade = len(df[(df['da']==False) & (df['ds']==False) & (df['de']==False)])
perc = grade/len(df)
print (f"Не распознано по специализации {grade} cообщений. Что составляет {perc:.1%}")
```

```python
df[(df['da']==False) & (df['ds']==False) & (df['de']==False)].head(10)
```

- Сообщения, которые не прошли разделение по специализации (DA,DE,DS), в основном содержат информацию о переписках, подкастах, новостях,поиске преподавателей в сфере IT. Для нашего исследования они не пригодны , поэтому мы можем удалить нерелевантную информацию 

```python
df = df.drop(df[(df['da']==False) & (df['ds']==False) & (df['de']==False)].index)
```

```python
def get_keyword_df(df, keywords):
    for keyword, keywords_to_search in keywords.items():
        keywords_regex = '|'.join(keywords_to_search)
        df[keyword] = df['text'].str.contains(keywords_regex, regex=True)
    keywords_df = pd.DataFrame(columns=['keyword', 'number'])
    for keyword in list(keywords.keys()):
        keywords_df = keywords_df.append({'keyword': keyword, 'number': len(df[df[keyword]])}, ignore_index=True)
    keywords_df = keywords_df.sort_values(by=['number'], ascending=False).reset_index(drop=True)
    keywords_df['perc'] = keywords_df['number'] / len(df) * 100
    return keywords_df
```

```python
# составим регулярку для отбора номера телефона
phone_number_regexp = r'([\+]?[78][ \-\(]*\d{3}[ \)\-]*\d{3}[ \-\(]*\d{2}[ \-\(]*\d{2})'
link_regexp = r'<.*?>'
# регулярка для даты формата YYYY-MM-DD и DD-MM-YYYY (также подойдёт YYYY.MM.DD и DD.MM.YYYY)
date_regexp = r'(\d{4}[\.\- ]*\d{2}[\.\- ]*\d{2})|(\d{2}[\.\- ]*\d{2}[\.\- ]*\d{4})'

# регулярка для времени (найдёт <<13:00>>, <<13-00>>, <<17 - 00>>, <<15 : 45>>, при условии что перед ними не было чисел)
time_regexp = r'(\D[123456789]\d[ ]*[:-]+[ ]*[123456789]\d\D)|(\D\d{2}[ \:\-]+0{2}\D)'

# для почты
email_regexp = r'\w*@\w+\.\w+'

# регулярка которая будет находить формат <число название_месяца> (Найдёт : 10 сент, не найдёт - 88 дек)
date_1_regexpr = r'(\D{1}\d{1} |[12]\d{1} |3[01]{1} )(сент|окт|ноя|дек|янв|фев|март|апр|май|июн|июл|мая|авг)'
```

```python
# чистим текст
df['clean_text'] = df['text']
df['clean_text'] = df['clean_text'].str.replace(phone_number_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(link_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(date_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(time_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(email_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(date_1_regexpr, ' ')
```

```python
# регулярное выражение для зп
number = r'([123456789]\d{0,2}[,\'\. ]?\d0{2}|[123456789]\d{2}0{1,2}|[123456789]\d{1,2}|[123456789]\d{0,1}[\.,\']\d{1,3})'

# для выражений которые стоят между числами з/п если они относятся к одной ставке
middle = r'[ \t]*[^0123456789\.,\'%]{0,9}?(?:-|до|–|—|to|-)[ \t]*[^0123456789\.,\'%]{0,1}?'

# регулярка для выражений, которые часто стоят перед зп
start = r'(?:net|gross|гросс|нет|rub|$|€|£|eur|dol|mid|sen|jun|мидл|синь|джун|зп|з\\п|плат|компенсац|' + \
        r'вилк|з\/п|доход|fork|moneys|salary|moneyparrot|деньг|оклад|условия|ставка|до|_plug).{0,10}?\s*?'

# регулярка для выражений которые часто стоят после зп
end = r'[ \t]*?[^0123456789]{0,3}?(?:k |к |тыс\.|т\.р|net|gross|гросс|нет|rub|$|€|£|eur|dol|mid|' + \
      r'sen|jun|мидл|синь|jun|р |джун|руб)'
```

```python
pattern1 = number + middle + number + end
pattern2 = start + number + end
pattern3 = number + end
pattern4 = start + number
df['salary'] = -1

df['salary'] = df['clean_text'].str.findall(pattern1)
df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern2)

df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern3)

df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern4)
```

```python
# все диапазоны з/п записались в tuple. Также если у нас было несколько диапазонов, то несколько tuple в одном листе
# распакуем и приведём к виду : 1 вакансия ~ среднее от всех указанных з/п 
def from_tuple_to_list(column):
    res = []
    for j in range(len(column)):      
        listr = []
        if type(column[j][0]) is tuple:
            for el in column[j]:
                listr += [*el]   
        elif type(column[j][0]) is str:
            listr += column[j]
        res.append(listr)
    return res

df.loc[df['salary'].str.len() != 0, 'salary'] = \
           from_tuple_to_list(column = df[df['salary'].str.len() != 0]['salary'].reset_index(drop = True))
```

```python
# возьмём среднее для диапазона
result_salary = df[df['salary'].str.len() != 0]['salary'].apply(lambda x : \
                                        np.array([float(m
                                                        .replace(',', '.') # 4,5 -> 4.5 
                                                        .replace(' ', '') # 50 000 -> 50000
                                                       ) for m in x]).mean())

df.loc[df['salary'].str.len() != 0, 'salary'] = result_salary
df.loc[df['salary'].str.len() == 0, 'salary'] = -1
```

```python
def Find(string):
    ''' Функция для поиска в тексте ссылки https'''
  
    # findall() использовался с допустимыми условиями для URL-адресов в строке 
    regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))"
    url = re.findall(regex,string)      
    return (" ".join(map(str,[x[0] for x in url])))
     #[x[0] for x in url] 
           

# проверка работы функции
string = df['text'][13]
print("Urls: ", Find(string))

# применение функции к новому столбцу датафрейма 
df['URLS'] = df['text'].apply(Find)
```

```python
def grade_category(df, category):
    """Функция  определяющая количество вакансий по специализации и градации"""
    level = {
        'Junior': df[(df[category] == 1) & (df['junior']==1)].junior.count(),
        'Middle': df[(df[category] == 1) & (df['middle']==1)].middle.count(),
        'Senior': df[(df[category] == 1) & (df['senior']==1)].senior.count(),
        'Lead': df[(df[category] == 1) & (df['lead']==1)].lead.count(),
        'Intern': df[(df[category] == 1) & (df['intern']==1)].intern.count()
    }
    return level
```

```python
def education_category (df, education):
    """ Функция определяющая количество вакансий с наличием профильного образованию у кандидатов"""
    level = {
        'Data Analyst': df[(df[education] == 1) & (df['da']==1)].da.count(),
        'Data Engineer': df[(df[education] == 1) & (df['de']==1)].de.count(),
        'Data Science': df[(df[education] == 1) & (df['ds']==1)].ds.count()
    }
    return level        
```

## Проверка гипотез 

 
  - Средняя зарплата  по рунку труда у Data Science грейда Middle и у Data Analyst Middle различаются
  - На рынке труда больше заявлено вакансий грейда Middle специалистов, чем Junior 
  

```python
def df_fitches (df,da,junior,rub,a, bins):
    """ Функция принимающая на вход признаки 
    (датафрейм, специализация, грейд, зарплата в валюте, ограничения по заработной плате),
    визуализирует распределение по данным параметрам  """
    df_fitches = (df[(df[da]==1) & (df[junior]==1) & (df[rub]==1) & (df['salary']>a)]['salary']
                  .reset_index().drop('index', axis =1))
    ax =sns.histplot(df_fitches, x = 'salary', kde = True, bins = bins);
    return ax
```

```python
plt.figure(figsize=(8, 5))
ax1 = df_fitches(df,'ds','middle','rub',15000,20)
plt.title('Распределение заработной платы Data Science Middle (руб)');

plt.figure(figsize=(8, 5))
ax2 = df_fitches(df,'ds','middle','dol',300,10)
plt.title('Распределение заработной платы Data Science Middle ($)');

plt.figure(figsize=(8, 5))
ax3 = df_fitches(df,'ds','middle','dol',300,10)
plt.title('Распределение заработной платы Data Science Middle ($)');
```

```python
plt.figure(figsize=(8, 5))
ax1 = df_fitches(df,'da','middle','rub',15000,20)
plt.title('Распределение заработной платы Data Аnalyst Middle (руб)');

plt.figure(figsize=(8, 5))
ax2 = df_fitches(df,'da','middle','dol',300,10)
plt.title('Распределение заработной платы Data Аnalyst Middle ($)');

plt.figure(figsize=(8, 5))
ax3 = df_fitches(df,'da','middle','eur',300,10)
plt.title('Распределение заработной платы Data Аnalyst Middle ($)');
```

Проверка гипотез будет осуществлять по заработнгой плате в рублях.

**H0: заработная плата у специалистов Data Analyst Middle и Data Science Middle не различается**

**H1: заработная плата у специалистов Data Analyst Middle и Data Science Middle различается**

Будем использовать t-критерий Стьюдента, так как сравниваемые выборки независимы между собой.

Примем alpha = .05 - критический уровень статистической значимости. Если p-value окажется меньше него - отвергаем нулевую  гипотезу.

Используем гипотезу о равенстве средних двух генеральных совокупностей.

```python
# применяем t тест для средних двух генер.совокупн
alpha =0.05    

results = st.ttest_ind(df[(df['da']==1) & (df['middle']==1) & (df['rub']==1) & (df['salary']>15000)]['salary'], 
                       df[(df['ds']==1) & (df['middle']==1) & (df['rub']==1) & (df['salary']>15000)]['salary'],
equal_var = False)     #дисперсии в выборках различны, equal_var = False

print('p-значение:', results.pvalue)                             

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```
