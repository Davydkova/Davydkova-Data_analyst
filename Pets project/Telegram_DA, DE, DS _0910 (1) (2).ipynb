```python
def get_keyword_df(df, keywords):
    for keyword, keywords_to_search in keywords.items():
        keywords_regex = '|'.join(keywords_to_search)
        df[keyword] = df['text'].str.contains(keywords_regex, regex=True)
    keywords_df = pd.DataFrame(columns=['keyword', 'number'])
    for keyword in list(keywords.keys()):
        keywords_df = keywords_df.append({'keyword': keyword, 'number': len(df[df[keyword]])}, ignore_index=True)
    keywords_df = keywords_df.sort_values(by=['number'], ascending=False).reset_index(drop=True)
    keywords_df['perc'] = keywords_df['number'] / len(df) * 100
    return keywords_df
```

```python
# —Å–æ—Å—Ç–∞–≤–∏–º —Ä–µ–≥—É–ª—è—Ä–∫—É –¥–ª—è –æ—Ç–±–æ—Ä–∞ –Ω–æ–º–µ—Ä–∞ —Ç–µ–ª–µ—Ñ–æ–Ω–∞
phone_number_regexp = r'([\+]?[78][ \-\(]*\d{3}[ \)\-]*\d{3}[ \-\(]*\d{2}[ \-\(]*\d{2})'
link_regexp = r'<.*?>'
# —Ä–µ–≥—É–ª—è—Ä–∫–∞ –¥–ª—è –¥–∞—Ç—ã —Ñ–æ—Ä–º–∞—Ç–∞ YYYY-MM-DD –∏ DD-MM-YYYY (—Ç–∞–∫–∂–µ –ø–æ–¥–æ–π–¥—ë—Ç YYYY.MM.DD –∏ DD.MM.YYYY)
date_regexp = r'(\d{4}[\.\- ]*\d{2}[\.\- ]*\d{2})|(\d{2}[\.\- ]*\d{2}[\.\- ]*\d{4})'

# —Ä–µ–≥—É–ª—è—Ä–∫–∞ –¥–ª—è –≤—Ä–µ–º–µ–Ω–∏ (–Ω–∞–π–¥—ë—Ç <<13:00>>, <<13-00>>, <<17 - 00>>, <<15 : 45>>, –ø—Ä–∏ —É—Å–ª–æ–≤–∏–∏ —á—Ç–æ –ø–µ—Ä–µ–¥ –Ω–∏–º–∏ –Ω–µ –±—ã–ª–æ —á–∏—Å–µ–ª)
time_regexp = r'(\D[123456789]\d[ ]*[:-]+[ ]*[123456789]\d\D)|(\D\d{2}[ \:\-]+0{2}\D)'

# –¥–ª—è –ø–æ—á—Ç—ã
email_regexp = r'\w*@\w+\.\w+'

# —Ä–µ–≥—É–ª—è—Ä–∫–∞ –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å —Ñ–æ—Ä–º–∞—Ç <—á–∏—Å–ª–æ –Ω–∞–∑–≤–∞–Ω–∏–µ_–º–µ—Å—è—Ü–∞> (–ù–∞–π–¥—ë—Ç : 10 —Å–µ–Ω—Ç, –Ω–µ –Ω–∞–π–¥—ë—Ç - 88 –¥–µ–∫)
date_1_regexpr = r'(\D{1}\d{1} |[12]\d{1} |3[01]{1} )(—Å–µ–Ω—Ç|–æ–∫—Ç|–Ω–æ—è|–¥–µ–∫|—è–Ω–≤|—Ñ–µ–≤|–º–∞—Ä—Ç|–∞–ø—Ä|–º–∞–π|–∏—é–Ω|–∏—é–ª|–º–∞—è|–∞–≤–≥)'
```

```python
# —á–∏—Å—Ç–∏–º —Ç–µ–∫—Å—Ç
df['clean_text'] = df['text']
df['clean_text'] = df['clean_text'].str.replace(phone_number_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(link_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(date_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(time_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(email_regexp, ' ')
df['clean_text'] = df['clean_text'].str.replace(date_1_regexpr, ' ')
```

```python
# —Ä–µ–≥—É–ª—è—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –∑–ø
number = r'([123456789]\d{0,2}[,\'\. ]?\d0{2}|[123456789]\d{2}0{1,2}|[123456789]\d{1,2}|[123456789]\d{0,1}[\.,\']\d{1,3})'

# –¥–ª—è –≤—ã—Ä–∞–∂–µ–Ω–∏–π –∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–æ—è—Ç –º–µ–∂–¥—É —á–∏—Å–ª–∞–º–∏ –∑/–ø –µ—Å–ª–∏ –æ–Ω–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –æ–¥–Ω–æ–π —Å—Ç–∞–≤–∫–µ
middle = r'[ \t]*[^0123456789\.,\'%]{0,9}?(?:-|–¥–æ|‚Äì|‚Äî|to|-)[ \t]*[^0123456789\.,\'%]{0,1}?'

# —Ä–µ–≥—É–ª—è—Ä–∫–∞ –¥–ª—è –≤—ã—Ä–∞–∂–µ–Ω–∏–π, –∫–æ—Ç–æ—Ä—ã–µ —á–∞—Å—Ç–æ —Å—Ç–æ—è—Ç –ø–µ—Ä–µ–¥ –∑–ø
start = r'(?:net|gross|–≥—Ä–æ—Å—Å|–Ω–µ—Ç|rub|$|‚Ç¨|¬£|eur|dol|mid|sen|jun|–º–∏–¥–ª|—Å–∏–Ω—å|–¥–∂—É–Ω|–∑–ø|–∑\\–ø|–ø–ª–∞—Ç|–∫–æ–º–ø–µ–Ω—Å–∞—Ü|' + \
        r'–≤–∏–ª–∫|–∑\/–ø|–¥–æ—Ö–æ–¥|fork|moneys|salary|moneyparrot|–¥–µ–Ω—å–≥|–æ–∫–ª–∞–¥|—É—Å–ª–æ–≤–∏—è|—Å—Ç–∞–≤–∫–∞|–¥–æ|_plug).{0,10}?\s*?'

# —Ä–µ–≥—É–ª—è—Ä–∫–∞ –¥–ª—è –≤—ã—Ä–∞–∂–µ–Ω–∏–π –∫–æ—Ç–æ—Ä—ã–µ —á–∞—Å—Ç–æ —Å—Ç–æ—è—Ç –ø–æ—Å–ª–µ –∑–ø
end = r'[ \t]*?[^0123456789]{0,3}?(?:k |–∫ |—Ç—ã—Å\.|—Ç\.—Ä|net|gross|–≥—Ä–æ—Å—Å|–Ω–µ—Ç|rub|$|‚Ç¨|¬£|eur|dol|mid|' + \
      r'sen|jun|–º–∏–¥–ª|—Å–∏–Ω—å|jun|—Ä |–¥–∂—É–Ω|—Ä—É–±)'
```

```python
pattern1 = number + middle + number + end
pattern2 = start + number + end
pattern3 = number + end
pattern4 = start + number
df['salary'] = -1

df['salary'] = df['clean_text'].str.findall(pattern1)
df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern2)

df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern3)

df.loc[df['salary'].str.len() == 0, 'salary'] = \
df.loc[df['salary'].str.len() == 0,'clean_text'].str.findall(pattern4)
```

```python
# –≤—Å–µ –¥–∏–∞–ø–∞–∑–æ–Ω—ã –∑/–ø –∑–∞–ø–∏—Å–∞–ª–∏—Å—å –≤ tuple. –¢–∞–∫–∂–µ –µ—Å–ª–∏ —É –Ω–∞—Å –±—ã–ª–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤, —Ç–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ tuple –≤ –æ–¥–Ω–æ–º –ª–∏—Å—Ç–µ
# —Ä–∞—Å–ø–∞–∫—É–µ–º –∏ –ø—Ä–∏–≤–µ–¥—ë–º –∫ –≤–∏–¥—É : 1 –≤–∞–∫–∞–Ω—Å–∏—è ~ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç –≤—Å–µ—Ö —É–∫–∞–∑–∞–Ω–Ω—ã—Ö –∑/–ø 
def from_tuple_to_list(column):
    res = []
    for j in range(len(column)):      
        listr = []
        if type(column[j][0]) is tuple:
            for el in column[j]:
                listr += [*el]   
        elif type(column[j][0]) is str:
            listr += column[j]
        res.append(listr)
    return res

df.loc[df['salary'].str.len() != 0, 'salary'] = \
           from_tuple_to_list(column = df[df['salary'].str.len() != 0]['salary'].reset_index(drop = True))
```

```python
# –≤–æ–∑—å–º—ë–º —Å—Ä–µ–¥–Ω–µ–µ –¥–ª—è –¥–∏–∞–ø–∞–∑–æ–Ω–∞
result_salary = df[df['salary'].str.len() != 0]['salary'].apply(lambda x : \
                                        np.array([float(m
                                                        .replace(',', '.') # 4,5 -> 4.5 
                                                        .replace(' ', '') # 50 000 -> 50000
                                                       ) for m in x]).mean())

df.loc[df['salary'].str.len() != 0, 'salary'] = result_salary
df.loc[df['salary'].str.len() == 0, 'salary'] = -1
```

```python
def Find(string):
    ''' –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–∏—Å–∫–∞ –≤ —Ç–µ–∫—Å—Ç–µ —Å—Å—ã–ª–∫–∏ https'''
  
    # findall() –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–ª—Å—è —Å –¥–æ–ø—É—Å—Ç–∏–º—ã–º–∏ —É—Å–ª–æ–≤–∏—è–º–∏ –¥–ª—è URL-–∞–¥—Ä–µ—Å–æ–≤ –≤ —Å—Ç—Ä–æ–∫–µ 
    regex = r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?¬´¬ª‚Äú‚Äù‚Äò‚Äô]))"
    url = re.findall(regex,string)      
    return (" ".join(map(str,[x[0] for x in url])))
     #[x[0] for x in url] 
           

# –ø—Ä–æ–≤–µ—Ä–∫–∞ —Ä–∞–±–æ—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏
string = df['text'][13]
print("Urls: ", Find(string))

# –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∫ –Ω–æ–≤–æ–º—É —Å—Ç–æ–ª–±—Ü—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ 
df['URLS'] = df['text'].apply(Find)
```

```python
# –∑–∞–º–µ–Ω–∞ —Ç–∏–ø–∞ —Å—Ç–æ–ª–±—Ü–∞ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–∏—Ö –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–π
df['URLS'] = df['URLS'].astype(str)

df['URLS'] = df ['URLS'].replace("['",' ')
df['URLS'] = df ['URLS'].replace("']",' ')
```

```python
def grade_category(df, category):
    """–§—É–Ω–∫—Ü–∏—è  –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π –ø–æ —Å–ø–µ—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –∏ –≥—Ä–∞–¥–∞—Ü–∏–∏"""
    level = {
        'Junior': df[(df[category] == 1) & (df['junior']==1)].junior.count(),
        'Middle': df[(df[category] == 1) & (df['middle']==1)].middle.count(),
        'Senior': df[(df[category] == 1) & (df['senior']==1)].senior.count(),
        'Lead': df[(df[category] == 1) & (df['lead']==1)].lead.count(),
        'Intern': df[(df[category] == 1) & (df['intern']==1)].intern.count()
    }
    return level
```

```python
def recolate_category(df,relocate):
    """ –§—É–Ω–∫—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π —Å —Ä–∞–∑–Ω—ã–º —Ñ–æ—Ä–º–∞—Ç–æ–º —Ä–∞–±–æ—Ç—ã –ø–æ —Å–ø–µ—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –∑–∞ 2022 –≥–æ–¥"""
    level= {
        'Data Analyst':df[(df[relocate]==1) & (df['da']==1) & (df['year']==2022)].da.count(),
        'Data Engineer':df[(df[relocate]==1) & (df['de']==1) & (df['year']==2022)].ds.count(),
        'Data Science':df[(df[relocate]==1) & (df['ds']==1) & (df['year']==2022)].de.count() 
    }
    return level      
```

```python
def education_category (df, education):
    """ –§—É–Ω–∫—Ü–∏—è –æ–ø—Ä–µ–¥–µ–ª—è—é—â–∞—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–∞–∫–∞–Ω—Å–∏–π —Å –Ω–∞–ª–∏—á–∏–µ–º –ø—Ä–æ—Ñ–∏–ª—å–Ω–æ–≥–æ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—é —É –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤"""
    level = {
        'Data Analyst': df[(df[education] == 1) & (df['da']==1)].da.count(),
        'Data Engineer': df[(df[education] == 1) & (df['de']==1)].de.count(),
        'Data Science': df[(df[education] == 1) & (df['ds']==1)].ds.count()
    }
    return level        
```

```python
def df_fitches (df,da,junior,rub,a, bins):
    """ –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞—é—â–∞—è –Ω–∞ –≤—Ö–æ–¥ –ø—Ä–∏–∑–Ω–∞–∫–∏ 
    (–¥–∞—Ç–∞—Ñ—Ä–µ–π–º, —Å–ø–µ—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è, –≥—Ä–µ–π–¥, –∑–∞—Ä–ø–ª–∞—Ç–∞ –≤ –≤–∞–ª—é—Ç–µ, –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –ø–æ –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç–µ),
    –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ –¥–∞–Ω–Ω—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º  """
    df_fitches = (df[(df[da]==1) & (df[junior]==1) & (df[rub]==1) & (df['salary']>a)]['salary']
                  .reset_index().drop('index', axis =1))
    ax =sns.histplot(df_fitches, x = 'salary', kde = True, bins = bins);
    return ax
```

**H0: –∑–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞ —É —Å–ø–µ—Ü–∏–∞–ª–∏—Å—Ç–æ–≤ Data Analyst Middle –∏ Data Science Middle –Ω–µ —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è**

**H1: –∑–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞ —É —Å–ø–µ—Ü–∏–∞–ª–∏—Å—Ç–æ–≤ Data Analyst Middle –∏ Data Science Middle —Ä–∞–∑–ª–∏—á–∞–µ—Ç—Å—è**

–ë—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å t-–∫—Ä–∏—Ç–µ—Ä–∏–π –°—Ç—å—é–¥–µ–Ω—Ç–∞, —Ç–∞–∫ –∫–∞–∫ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º—ã–µ –≤—ã–±–æ—Ä–∫–∏ –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã –º–µ–∂–¥—É —Å–æ–±–æ–π.

–ü—Ä–∏–º–µ–º alpha = .05 - –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏. –ï—Å–ª–∏ p-value –æ–∫–∞–∂–µ—Ç—Å—è –º–µ–Ω—å—à–µ –Ω–µ–≥–æ - –æ—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é  –≥–∏–ø–æ—Ç–µ–∑—É.

–ò—Å–ø–æ–ª—å–∑—É–µ–º –≥–∏–ø–æ—Ç–µ–∑—É –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≥–µ–Ω–µ—Ä–∞–ª—å–Ω—ã—Ö —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–µ–π.

```python
# –ø—Ä–∏–º–µ–Ω—è–µ–º t —Ç–µ—Å—Ç –¥–ª—è —Å—Ä–µ–¥–Ω–∏—Ö –¥–≤—É—Ö –≥–µ–Ω–µ—Ä.—Å–æ–≤–æ–∫—É–ø–Ω
alpha =0.05    

results = st.ttest_ind(df[(df['da']==1) & (df['middle']==1) & (df['rub']==1) & (df['salary']>15000)]['salary'], 
                       df[(df['ds']==1) & (df['middle']==1) & (df['rub']==1) & (df['salary']>15000)]['salary'],
equal_var = False)     #–¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤ –≤—ã–±–æ—Ä–∫–∞—Ö —Ä–∞–∑–ª–∏—á–Ω—ã, equal_var = False

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)                             

if (results.pvalue < alpha):
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
```
