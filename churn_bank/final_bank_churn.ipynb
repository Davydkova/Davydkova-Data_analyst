## Проверка статистических гипотез 


### Гипотеза о различии доходов у клиентов в оттоке / не в оттоке

Сформулируем гипотезу:

H0: среднее значение дохода клиентов, ушедших в отток равно среднему значению дохода клиентов, оставшихся с банком

H1: среднее значение дохода клиентов, ушедших в отток отличается от среднего значению дохода клиентов, оставшихся с банком


Для анализа стат. теста необходимо проверить распределение признака на нормальность. Ранее, мы уже наблюдали в грубом приближении раввномерное распределение признака . Проверим еще раз распределение выборки зарплаты клиентов в разрезе оттока.

```python
#выборки в разрезе оттока клиентов

sample_0=df.query("churn==0")
sample_1=df.query("churn==1")
```

```python
def disp_col (df,i):
    ''' Функция проверки распределения выборки на нормальность при оттоке/не оттоке клиентов'''
    sns.displot(data=df,x=i,hue='churn', kde=True, multiple="stack")
    plt.title('Распределение параметра',fontsize=16)
    plt.show()
    
```

```python
disp_col(df,'estimated_salary')
```

Для оценки стат.гипотез мы можем применить Критерий t Стьюдента. Критерий направлен на оценку различий величин средних значений двух выборок X и Y, которые распределены по нормальному закону.

Также имеет значение равенство дисперсий (распределения) сравниваемых групп. При неравных дисперсиях применяется t-критерий в модификации Уэлча.

```python
def  col_var(i):
    print('Дисперсия ушедних {:.2f}'.format(sample_0[i].var()))
    print('Дисперсия оставшихся {:.2f}'.format(sample_1[i].var()))
```

```python
col_var('estimated_salary')
```

```python
#пороговое значения уровня стат.значимости
alpha=0.05 

def t_test(sample1,sample2, i, a):
    '''Функция которая проверяет стат.значимость входных выборок на основании t теста (тест Стьюдента)''' 
    results = st.ttest_ind(
    sample1[i], 
    sample2[i], equal_var=a)
   
    print('p-значение:', results.pvalue)
    if (results.pvalue < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

```python
t_test(sample_0,sample_1,'estimated_salary', False)
```

<div style="border:solid black 1px; padding: 20px">   
Нулевая гипотеза не отвергнута, т.е. статистически значимых различий между средними двух выборок не обнаружено.

### Гипотеза о различии наличий кредитной карты, которыми пользуется клиент, которые ушли и теми,которые остались

Выделим две контрольные группы: налиичие кредитной карты у клиентов (sample_A) и наличие кредитной карты у клиентов в оттоке (sample_B). Проверим находят ли статистические критерии (разница средних) разницу между выборками А и B.

```python
# сформируем выборки

sample_A = list(df.groupby('credit_card').agg({'user_id':'count'})['user_id'])
sample_B = list(df.query('churn == 1').groupby('credit_card').agg({'user_id':'count'})['user_id'])
```

```python
# напишем функцию для проверки равенства долей
def z_test(before, after):
    successes = np.array(after)
    trials = np.array(before)
     
    p1 = successes[0]/trials[0]
    p2 = successes[1]/trials[1]
    
    p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

    difference = p1 - p2

    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

    distr = st.norm(0, 1) 

    p_value = (1 - distr.cdf(abs(z_value))) * 2

    return p_value
```

```python
# получаем значение p-value
p_value = z_test(
    sample_A,
    sample_B
)

# сравниваем полученное значение с критерием
if p_value > alpha:
    print('Не получилось отвергнуть нулевую гипотезу')
else:
    print('Отвергаем нулевую гипотезу')
```

<div style="border:solid black 1px; padding: 20px">  
Нулевая гипотеза не отвергнута, т.е. статистически значимых различий между клиентами с наличием кредитной карты не обнаружено

## Кластеризация. Портрет пользователей склонных к оттоку.
